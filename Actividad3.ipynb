```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .csv sin indice
Credicel = pd.read_excel('PrepCrit3.xlsx')
```

```python
# Separamos los datos cuantitativos también
CredicelNum = Credicel.select_dtypes(np.number)
CredicelNum
```

```python
# Encontramos todas las correlaciones entre las variables

Corr_Factors = CredicelNum.corr()
Corr_Factors
```

```python
# Se aplica valor absoluto

Corr_Factors1 = abs(Corr_Factors)
Corr_Factors1
```

```python
# Graficamos el mapa de calor de los coeficientes de correlación

Heat_Map1 = sns.heatmap(Corr_Factors1, cmap = 'YlGnBu', annot = True, fmt = ".2f")
Heat_Map1

```

Regresión multiple Precio


```python
# Declaramos las variables dependientes e independientes para la regresión lineal

Vars_Indep = CredicelNum[['enganche', 'porc_enganche', 'costo_total', 'monto_financiado', 'semana']]
Var_Dep = CredicelNum['precio']
```

```python
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python

#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

Modelo matemático precio: 0.99386081x1 - 4.42372157x2 - 0.07511813x3 - 0.81253147x4 - 1.62462253x5 + (-142.65170863703997)


```python
# Corroboramos cual es el coeficiente de Determinación de nuestro modelo

coef_Deter = model.score(X = Vars_Indep, y = Var_Dep)
coef_Deter
```

```python
#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
